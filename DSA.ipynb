```python
# Here we create a node of double linked list. we also called it node of the class.
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
    self.prev=None

# Here we create a class of double linked list.
class DLL:
  def __init__(self):

    self.head=None

# Here we create a funcation Insert() this is use for insert the node at end of the double linked list.
  def Insert(self,data):
    new=Node(data)

    if self.head==None:
      self.head=new
      self.head.next=None
    else:
      temp=self.head
      while temp.next!=None:
        temp=temp.next
      temp.next=new
      new.prev=temp
      new.next=None

# Here we insert the node at the first position of the linked list.
  def Insert_at_1st(self,data):
    new=Node(data)
    new.next=self.head
    self.head.prev=new
    self.head=new

# Here we insert the node at any position of the linked list.

  def Insert_at_pos(self,pos,data):
    new=Node(data)
    temp=self.head

    if pos<1:
      print("Invalid position!!!")

    elif pos==1:
      new.next=self.head
      self.head.prev=new
      self.head=new

    else:
      try:
           for i in range(1,pos-1):
             temp=temp.next
           new.next=temp.next
           new.prev=temp
           new.prev.next=new
           new.next.prev=new

      except Exception:
         while temp.next!=None:
              temp=temp.next
         temp.next=new
         new.prev=temp
         new.next=None

# Delete at 1st position of double linked list.

  def delete_at_1st_pos(self):
    temp=self.head
    self.head=self.head.next
    self.head.prev=None
    temp=None


# Delete at last position of linked list.

  def delete_at_end(self):
    temp=self.head
    while temp.next.next!=None:
      temp=temp.next
    temp.next.prev=None
    temp.next=None

# Delete at any position of linked list.

  def delete_at_any_pos(self,pos):
    temp=self.head

    if pos<1:
      print("Invalid position!!!")
    elif pos==1:

       self.head=self.head.next
       self.head.prev=None
       temp=None

    else:
      try:

           for i in range(1,pos-1):
             temp=temp.next
           temp.next=temp.next.next
           temp.next.prev=temp.next

      except Exception:
         print("Invalid!!!")


# Here we create a Display() funcation. the main purpose of this function is to display the entire linked list.
  def Display(self):
    if self.head==None:
      print("Empty DLL!!!")
    else:
      temp=self.head
      while temp:
        print(temp.data,end="<-->")
        temp=temp.next

ob=DLL()
ob.Insert(10)
ob.Insert(20)
ob.Insert(30)
ob.Insert(40)
# ob.Insert_at_1st(100)
# ob.Insert_at_pos(6,56)
# ob.delete_at_1st_pos()
# ob.delete_at_end()
ob.delete_at_any_pos(5)
ob.Display()


```
